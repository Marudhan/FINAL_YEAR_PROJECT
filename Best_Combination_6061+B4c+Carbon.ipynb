```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# Sample dataset (Replace with real experimental data)
data = {
    'Composition': [
        "100% Al6061",
        "95% Al6061 + 5% B4C",
        "95% Al6061 + 5% Biochar",
        "95% Al6061 + 4% B4C + 1% Biochar",
        "90% Al6061 + 4% B4C + 6% Biochar"
    ],
    'Density (kg/m³)': [2670.15, 2628.41, 2575.97, 2618.28, 2571.86],
    'Youngs Modulus (GPa)': [69.91, 81.07, 72.26, 83.00, 80.53],
    'Thermal Conductivity (W/mK)': [163.25, 157.94, 152.70, 156.24, 150.72],
    'Tensile Strength (MPa)': [306.10, 328.60, 305.00, 331.90, 330.10],
    'Compression Strength (MPa)': [416.10, 438.60, 415.00, 441.90, 440.10],
    'Impact Strength (J)': [33.80, 36.41, 32.61, 36.86, 35.79],
    'Flexural Strength (MPa)': [496.10, 518.60, 495.00, 521.90, 520.10],
    'Hardness (BHN)': [75, 82, 78, 85, 80],  # Hypothetical values
    'Wear Resistance (mm³)': [0.15, 0.18, 0.14, 0.19, 0.17],  # Hypothetical values
    'Fatigue Strength (MPa)': [250, 270, 260, 280, 275]  # Hypothetical values
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display table in formatted style
print("\nFormatted Mechanical and Thermal Properties Table:\n")
print(df.to_string(index=False))

# Visualizing each mechanical test separately
properties = [
    'Tensile Strength (MPa)', 'Impact Strength (J)', 'Hardness (BHN)',
    'Flexural Strength (MPa)', 'Wear Resistance (mm³)', 'Fatigue Strength (MPa)'
]

for prop in properties:
    plt.figure(figsize=(8, 5))
    sns.barplot(x=df['Composition'], y=df[prop], palette='coolwarm')
    plt.xticks(rotation=45, ha='right')
    plt.xlabel("Composition")
    plt.ylabel(prop)
    plt.title(f"{prop} for Different Compositions")
    plt.show()

# Correlation Matrix Visualization
plt.figure(figsize=(12, 6))
sns.heatmap(df.drop(columns=['Composition']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Material Properties")
plt.show()

# Pairplot Visualization
sns.pairplot(df.drop(columns=['Composition']), diag_kind='kde')
plt.show()

# Boxplot Visualization
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns=['Composition']))
plt.xticks(rotation=45, ha='right')
plt.title("Boxplot of Mechanical Properties")
plt.show()

# Predicting the best combination of Al6061 + B4C + Biochar
X = df.drop(columns=['Composition'])
y_columns = list(X.columns)

best_combination = {}

for prop in y_columns:
    best_combination[prop] = df.loc[df[prop].idxmax(), 'Composition']

print("\nBest Combination for Each Property:\n")
for prop, comp in best_combination.items():
    print(f"{prop}: {comp}")
```
